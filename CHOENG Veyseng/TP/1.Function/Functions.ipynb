## Recursive Function

```python
def factorial(n):
    if n < 0:
        return "Error: only positive number"
    elif n == 0:
        return 1
    else:
        return n* factorial(n - 1)

print(factorial(9))
```

```python
def countdown_num(n):
    if n == 0:
        return 0
    elif n < 0:
        print(n)
        return countdown_num(n + 1)
    else:
        print(n)
        return countdown_num(n - 1)

num = int(input("Enter number: "))
print(countdown_num(num))
```

```python
def exp_num(n, m):
    if m == 0:
        return n
    else:
        return n * exp_num(n, m - 1)

print(exp_num(5,2))
```
