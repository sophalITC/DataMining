###**Reduce High Dimensionality**

There are many models that would be overfitted when there are high dimensionality of inputs.By reducing the dimensionality, it would provide some remarkable benifits such as:

*   Reduced training times
*   Simplified and interpretable models
*   Reduced chances of overfitting i.e. lesser variance
*   Less impact of the curse of dimensionality

###**Forward Feature Selection**
The purpose of doing forward feature selection is to arrange the features in descending order of their importance.

####How we filter the best hyperparameter?
*   We start by selecting one feature and calculating the metric value for each feature on cross-validation dataset
*   The feature offering best metric value is selected and appended to list of features
*   The process is reiterated, this time with two features, one selected from the previous iteration and the other one selected from the set of all features not present in the set of already chosen features
*   The metric value is computed for each set of 2 features and feature offering best metric value is appended to the list of relevant features
*   This process is repeated until we have the desired number of features (n in this case)

Source: https://towardsdatascience.com/feature-importance-and-forward-feature-selection-752638849962

```python
import pandas as pd
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
Data = pd.read_csv(file_path)
```

```python
X = Data.iloc[:,2:-1]
y = Data.iloc[:,1:2]
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state = 1, train_size = 0.8)
```

```python
def evaluate_metric(model, x_test, y_test):
    return f1_score(y_test, model.predict(x_test), average='micro')
```

```python
def forward_feature_selection(x_train, x_cv, y_train, y_cv, n):
    feature_set = []
    for num_features in range(n):
        metric_list = [] # Choose appropriate metric based on business problem
        model = SGDClassifier() # You can choose any model you like, this technique is model agnostic
        for feature in x_train.columns:
            if feature not in feature_set:
                f_set = feature_set.copy()
                f_set.append(feature)
                model.fit(x_train[f_set], y_train)
                metric_list.append((evaluate_metric(model, x_cv[f_set], y_cv), feature))

        metric_list.sort(key=lambda x : x[0], reverse = True) # In case metric follows "the more, the merrier"
        feature_set.append(metric_list[0][1])
    return feature_set
```

```python
f = forward_feature_selection(x_train, x_test, y_train, y_test, 10)
```
