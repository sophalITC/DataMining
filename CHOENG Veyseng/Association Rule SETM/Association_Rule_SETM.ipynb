```python
def read_csv(path):
  f = open(path, 'r')
  lines = f.readlines()
  f.close()
  x = []
  for line in lines:
    d = line.split(';')
    x.append((d[0], d[1]))
    
  return x
```

```python
datalist = read_csv('/content/drive/MyDrive/online_retail_II.csv')
datalist = datalist[1:]
print(len(datalist))
```

```python
def filter_transaction_id(data):
  L = []
  l = []
  i = 0
  while i < len(data):
    l.append(data[i][1])
    while i + 1 < len(data):
      if data[i][0] == data[i+1][0]:
        l.append(data[i+1][1])
        i += 1
      else:
        break
        
    L.append((data[i][0], l))
    l = []
    i += 1

  return L
```

```python
filter_datalist = filter_transaction_id(datalist)
```

```python
print(len(filter_datalist))
```

```python
def find_support(data):
  unique = []
  for i in range(len(data)):
    if data[i][1] not in unique:
      unique.append(data[i][1])

  support_dictionary = dict.fromkeys(unique, 0)
  for d in data:
    support_dictionary[d[1]] += 1
    
  return unique, support_dictionary
```

```python
items, support_item_dict = find_support(datalist)
```

```python
print(len(items), type(support_item_dict))
```

```python
def displaySupportL1(support):
  support = sorted(support.items(), key=lambda x: x[1], reverse=True)
  for n in support:
    key = n[0]
    value = n[1]
    print("--------------------------------------")
    print(f"Item: %s, Support: %d" % (key, value))
```

```python
displaySupportL1(support_item_dict)
```
