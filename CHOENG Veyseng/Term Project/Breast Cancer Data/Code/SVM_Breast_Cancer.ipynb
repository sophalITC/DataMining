```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
from sklearn import svm
import seaborn as sns
import pandas as pd
```

```python
#Preparing for data
bc = datasets.load_breast_cancer()
X, y = bc.data, bc.target

n_samples, n_features = X.shape

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 123)
```

```python
#Scalar
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(len(X_test[0]))
```

```python
#Model 1 using rbf kernel which is radial basis function
model1 = svm.SVC(kernel='rbf', C=3)
model1.fit(X_train, y_train)
```

```python
z_test = model1.predict(X_test)

n_correct = 0
for i in range(len(z_test)):
  if z_test[i] == y_test[i]:
    n_correct += 1

acc = (n_correct / len(z_test))
print(f"Accuracy = {acc:.4f}")
report = classification_report(y_test, z_test, labels=[0, 1], target_names=["Benign", "Malignant"], output_dict=True)
```

```python
sns.heatmap(pd.DataFrame(report).iloc[:-1, :].T, annot=True)
```

```python
#Model 1 using rbf kernel which is radial basis function
model2 = svm.SVC(kernel='linear', C=3)
model2.fit(X_train, y_train)
```

```python
z_test = model2.predict(X_test)

n_correct = 0
for i in range(len(z_test)):
  if z_test[i] == y_test[i]:
    n_correct += 1

acc = (n_correct / len(z_test))
print(f"Accuracy = {acc:.4f}")
report = classification_report(y_test, z_test, labels=[0, 1], target_names=["Benign", "Malignant"], output_dict=True)
```
