# <center>Backward Elimination Feature Selection</center>



<b>Feature selection </b> is a way of selecting the subset of the most relevant features from the original features set by removing the redundant, irrelevant, or noisy features.

Feature selection usually can lead to better learning performance, higher learning accuracy, lower computational cost, and better model interpretability. 

<b> Backward Elimination</b> s a feature selection technique while building a machine learning model. It is used to remove those features that do not have a significant effect on the dependent variable or prediction of output. 

```python
#importing the libraries
import pandas as pd
```

```python

#reading the file

data = pd.read_csv('Dataset.csv')
data.head()
```

```python
#shape of the data
data.shape
```

```python
# checking missing values in the data
data.isnull().sum()
```

```python
# creating the training data
X = data.drop(['Location', 'Owner_Type'], axis=1)
y = data['Owner_Type']
```

X here will be the independent variables after dropping the "Location" variable and y will be the target variable, “Owner_Type”. Let me print the shape of both of these-

```python
X.shape, y.shape
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LinearRegression
```

```python
lreg = LinearRegression()

sfs1 = sfs(lreg, k_features=6, forward=False, verbose=1, scoring='neg_mean_squared_error')
```

```python
sfs1 = sfs1.fit(X, y)
```

```python
feat_names = list(sfs1.k_feature_names_)
print(feat_names)
```

```python
new_data = data[feat_names]
new_data['Price'] = data['Price']

new_data.head()
```
