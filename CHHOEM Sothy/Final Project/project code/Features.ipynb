```python
#import Libraries
import pandas as pd
import glob
# visualization
import seaborn as sns
import matplotlib.pyplot as plt
# machine learning
import numpy as np
```

# read data 

```python
path ='dataset' # use your path
all_files = glob.glob(path + "/*.csv")

li = []
li1=[]

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df['company']=filename.split("/")[-1].split(".")[0]
    if('unclean' in filename ):
        df['tax']=0
        df['mpg']=0
        li1.append(df)
    else:
        if('hyundi' in filename):
            df.rename(columns={'tax(£)': 'tax'}, inplace=True)
        if ('focus' in filename or 'cclass' in filename):
            df['tax']=0
            df['mpg']=0
        li.append(df)
data = pd.concat(li, axis=0, ignore_index=True)
```

# Test model Regression for car prices

Test on 11 Modle in Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error 
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import median_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from sklearn import linear_model
from sklearn import metrics #, svm

RandomForestRegressor = RandomForestRegressor()
DecisionTreeRegressor = DecisionTreeRegressor()
MLPRegressor=MLPRegressor()
classifiers =[RandomForestRegressor,DecisionTreeRegressor,MLPRegressor]+[
    linear_model.BayesianRidge(),
    linear_model.LinearRegression(),
    linear_model.Ridge(),
    linear_model.ARDRegression(),
    linear_model.SGDRegressor(),
    linear_model.PassiveAggressiveRegressor(),
    #linear_model.LassoLars(),
    linear_model.TheilSenRegressor()
]
All_model_train=[]
for item in classifiers:
    print(item)
    test_model = item
    test_model.fit(X_train,y_train) 
    #Calculating Details
    print('Linear model Train Score is : ' , test_model.score(X_train, y_train))
    print('Linear model Test Score is  : ' , test_model.score(X_test, y_test))
    print('----------------------------------------------------')
    
    y_pred = test_model.predict(X_test)
    yt_pred = test_model.predict(X_train)
    print('Predicted Value for Linear Regression is : \n' ,np.around(y_pred[:10],3))
    print('real Value for Linear Regression is : \n' ,np.around(np.array(y_test[:10]),3))
    #----------------------------------------------------
    plt.figure(figsize=(10,6))
    plt.scatter(yt_pred,yt_pred - y_train,c = 'black', marker = 'o', s = 35, alpha = 0.5,label = 'Train data')
    plt.scatter(y_pred,y_pred - y_test, c = 'c', marker = 'o', s = 35, alpha = 0.7,label = 'Test data')
    plt.xlabel('Predicted values')
    plt.ylabel('Tailings')
    plt.legend(loc = 'upper left')
    plt.show()#Calculating Mean Absolute Error
    MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
    print('Mean Absolute Error Value is : ', MAEValue)
    #----------------------------------------------------
    #Calculating Mean Squared Error (the average of the squares of the errors)
    MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
    print('Mean Squared Error Value is  : ', MSEValue)
    #----------------------------------------------------
    #Calculating Median Squared Error
    MdSEValue = median_absolute_error(y_test, y_pred)
    print('Median Squared Error Value is : ', MdSEValue )
```
