```python
import requests
from bs4 import BeautifulSoup as bs

# load the projectpro webpage content
r = requests.get('https://www.slideshare.net/tusharpanda88/python-basics-59573634?qid=ab2be908-b93a-4d87-ab8b-62ed06f3c3f2&v=&b=&from_search=4')

# convert to beautiful soup
soup = bs(r.content)

# printing our web page
print(soup.prettify())
```

```python
#creat new folder MyImage
import os
dir = os.path.join("MyImages")
if not os.path.exists(dir):
    os.mkdir(dir)
#delete all file in MyImage
import os, shutil
folder = 'MyImages'
for filename in os.listdir(folder):
    file_path = os.path.join(folder, filename)
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
            os.unlink(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
    except Exception as e:
        print('Failed to delete %s. Reason: %s' % (file_path, e))
```

```python
from pathlib import Path
save_location = Path("MyImages")
save_location.mkdir(exist_ok=True) # create the directory if it does not exist already
links = soup.find_all("div", {'class': 'slide'})
filename=[]
print(links)
for link in links :
    #print("ok")
    for img in link.find_all('img', srcset=True):
  
        img = img["srcset"]
     
        imgString= img
        t = imgString.rfind(',')
        t1 = imgString[t+1:]
        tt = t1.rfind('1024w')
        img= t1[0:tt]
        print(img)
        
        
        #if img.endswith('.jpg'): 
        #    print("ok")
        myString= img
        a = myString.rfind('/')
        cut1 = myString[a+1:]
        b = cut1.rfind('?')
        name= cut1[0:b]
        #print(cut1)
        #print(filename)
        filename.append(name)
        
        #///////////////////////// 
        #download 
        img_data = requests.get(img).content
        
        with open(save_location / name, 'wb') as handler:
            handler.write(img_data)
print(img_data)
```

```python
import cv2
import os
from PIL import Image

def load_images_from_folder(folder):
    images = []
    page=0
    ii=[]
    iii=[]
    for i in os.listdir(folder):
        
        if(page==0):
            z=Image.open(r''+folder+'/'+i)
            first = z.convert('RGB')
            page +=1
            name=i
        if(page>0):
            im=Image.open(r''+folder+'/'+i)
            im = im.convert('RGB')
            images.append(im)
     
    return name,first , images
name,first,all_img=load_images_from_folder('MyImages')
n = name.rfind('.')
name= name[0:n]
all_img=all_img[1:n]
```
