## Median

Median is the value in the middle, after we have sorted values ascending.
#### Note that imputing missing data with median value can only be done with numerical data.

First, we need to import some important library such as Numpy, Pandas, and so on.
Then we have to read file and 

```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

#### Check which field has missing value

```python
# read file
df=pd.read_csv("train.csv").iloc[:,1:]
print('File read succesfully !', f'Shape of original file: {df.shape}')
# count the number of NaN values in each column
print(df.isnull().sum())
```

#### Fill missing value using Median function

```python
X = df["Age"].median()
df["Age"].fillna(X, inplace=True)
df
```

#### Cheack missing value after filling

```python
# count the number of NaN values in each column
print(df.isnull().sum())
```

```python
df
```

#### Drop some fields that we don't work with them

```python
df.drop("Name",axis=1,inplace=True)
df.drop("Sex",axis=1,inplace=True)
df.drop("Ticket",axis=1,inplace=True)
df.drop("Cabin",axis=1,inplace=True)
df.drop("Embarked",axis=1,inplace=True)
df
```

```python
y = df['Survived']

X_train, X_test,y_train,y_test = train_test_split(df,y,test_size=0.3)
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred = lr.predict(X_test)
print('ACCURACY: %.2f%%' % (metrics.accuracy_score(pred,y_test)*100))
```

#### Write csv file after filling the missing value by using Median

```python
print("Write to output csv file ")
df.to_csv('output.csv')
```
