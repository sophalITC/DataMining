# Isolation Forest

Isolation Forest, randomly sub-sampled data is processed in a tree structure based on randomly selected features. The samples that travel deeper into the tree are less likely to be anomalies as they required more cuts to isolate them

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv('data_set_A.csv')

df.head()

x = df.values

plt.scatter(x[:,0], x[:,1])
```

```python
clf = IsolationForest(contamination=0.2)
clf.fit(x)
predictions = clf.predict(x)
```

```python
(predictions<0).mean()
```

```python
abn_ind = np.where(predictions < 0)
```

```python
plt.scatter(x[:,0], x[:,1])
plt.scatter(x[abn_ind,0], x[abn_ind,1], edgecolors="r")
```
