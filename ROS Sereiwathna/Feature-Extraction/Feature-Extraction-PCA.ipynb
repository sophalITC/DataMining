# Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is a dimensionality reduction algorithm. First it identifies the hyperplane that lies closest to the data, and then it projects the data onto it.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("/content/drive/MyDrive/M1_Semester2/DataMining/Feature-Extraction/cancer_tumor_data_features.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(data)
```

```python
pca = PCA(n_components=2)
```

```python
reduced_x = pca.fit_transform(x)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x[:,0], x[:,1]);
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(reduced_x[:,0], reduced_x[:,1]);
```
