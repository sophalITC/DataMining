## <center>What is Web Scraping?
Web scraping is a method of extracting data from web sites. It uses software to extract all the information available from the targeted site by simulating human behavior.

## <center>What is Bearutiful Soup?
Beautiful Soup is a Python library that makes it easy to scrape information from web pages. It sits atop an HTML or XML parser and provides Pythonic idioms for iterating, searching, and modifying the parse tree.

### Features
Some key features that make beautiful soup unique are:

    -Beautiful Soup provides a few simple methods and Pythonic idioms for navigating, searching, and modifying a 
    parse tree.
    -Beautiful Soup automatically converts incoming documents to Unicode and outgoing documents to UTF-8.
    -Beautiful Soup sits on top of popular Python parsers like lxml and html5lib, which allows​ us to try out different parsing strategies or trade speed for flexibility.

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
numOfPage = int(input("Please Enter Number of Page(s) to Scrap Data: "))
```

```python
p = 0
count = 0
allData_ = []
for page in range(1, numOfPage + 1):
    webpage = requests.get('https://www.realestate.com.kh/buy/?page='+str(page)).text
    soup = BeautifulSoup(webpage, 'lxml')
    # one data
    containers = soup.find_all('div', class_ = 'items-container')

    print(len(containers))
    
    p+=1
    print('Page', p, ': ')

    for sub_container in containers:
        sub_container1 = sub_container.find('div', class_ = 'items')
        
        for container in sub_container1:
            data_ = []
            count+=1
            print("count ",count,":")
            title_ = 'NaN'
            price_ = 'NaN'
            type_ = 'NaN'
            location_ = 'NaN'
            agency_ = 'NaN'
            numOfbedroom = 'NaN'
            numOfbathroom = 'NaN'
            carSpace = 'NaN'
            floorArea = 'NaN'
            landArea = 'NaN'
            floorLevel = 'NaN'
            facing = 'NaN'
            
            
            # find title
            try:
                title_con = container.find('header', class_= 'featured css-ineky e1jqslr40')
                title_ = title_con.find('a').text
            except AttributeError:
                title_con = container.find('header', class_= 'premium css-ineky e1jqslr40')
                title_ = title_con.find('a').text
            print(title_)
            
            # find price
            try:
                price_ = container.find('div', class_='price').text
            except AttributeError:
                print('')

            # find type
            try:
                type_ = container.find('div', class_= 'heading').text
                type_ = type_.replace(price_, '')
            except AttributeError:
                print('not found')
            print(type_)
             
            # find location
            location_ = container.find('div', class_= 'address').text[1:]
            print(location_)

            # find num of bedroom, num of bathroom, floor area, land area, floor level
            try:
                features = []
                features_con = container.find('div', class_ = 'private')
                features_ = features_con.find_all('span', class_ = 'label')
                for feature_ in features_:
                    features.append(feature_.text)
                for feature in features:
                    if 'Bedrooms: ' in feature:
                        numOfbedroom = feature.replace('Bedrooms: ', '')

                    if 'Bathrooms: ' in feature:
                        numOfbathroom = feature.replace('Bathrooms: ', '')

                    if 'Car spaces: ' in feature:
                        carSpace = feature.replace('Car spaces: ', '')

                    if 'Land area: ' in feature:
                        landArea = feature.replace('Land area: ', '')

                    if 'Floor level: ' in feature:
                        floorLevel = feature.replace('Floor level: ', '')

                    if 'Floor area: ' in feature:
                        floorArea = feature.replace('Floor area: ', '')

                    if 'Facing: ' in feature:
                        facing = feature.replace('Facing: ', '')
        
            except AttributeError:
                print('not found')
            print(numOfbedroom)
            print(numOfbathroom)
            print(carSpace)
            print(landArea)
            print(floorLevel)
            print(floorArea)
            print(facing)
            print(price_)

            # find agency
            agency_con = container.find('div', class_ = 'listed')
            agency_ = agency_con.find('span').text
            if(not str(agency_)):
                agency_ = 'NaN'
            print(agency_)

            # status
            status_ = container.find_all('div', class_ = 'translucent')
            for status in status_:
                print(status.text)
                
            data_.append(type_)
            data_.append(title_)
            data_.append(price_)
            data_.append(location_)
            data_.append(landArea)
            data_.append(numOfbedroom)
            data_.append(numOfbathroom)
            data_.append(carSpace)
            data_.append(floorLevel)
            data_.append(floorArea)
            data_.append(facing)
            data_.append(agency_)
            for s in status_:
                data_.append(s.text)

            allData_.append(data_)
            print(len(data_))
            print('\n')
```
