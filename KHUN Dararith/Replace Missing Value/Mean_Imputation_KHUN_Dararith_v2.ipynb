# <center>Mean Imputation
    
Missing values are common in dealing with real-world problems when the data is aggregated over long time streches form disparate sources and reliable machine learing modeling.
    
So today, i will talk about Imputation/Replacing Missing Values with Mean.

### Why do we use Mean Imputation technique?

We use this technique to estimate the missing values from the other training examples. This technique is used to replace all the missing values from dataset with mean value of the entire feature column. This technique can only be done with numerical data.

### How to implement this technique?


We first import some library such as, numpy and pandas

```python
import numpy as np
import pandas as pd

#df = pd.read_csv("/Users/dararithkhun/Downloads/Placement_Data_Full_Class.csv")
df = pd.read_csv("/Users/dararithkhun/OneDrive - College KH/Rith's_Mac/Downloads_Cloud/Telegram Desktop/handle-missing-values-master/test_dataset/train.csv")
df.head()

```

```python
df.isnull().sum()
```

```python
def detect_missing():
# checking missing values
    null_series = df.isnull().sum()
    print()
    null_column_list = []
    if sum(null_series):
        print('Following columns contains missing values : ')
        total_samples = df.shape[0]
        for i, j in null_series.items():
            if j:
                print("{} : {:.2f} %".format(i, (j/total_samples)*100))
                null_column_list.append(i)
    else:
        print("None of the columns contains missing values !")
    return null_column_list
```

```python
# printing missing detected values
null_column_list = detect_missing()
print("The Null Columns : ", null_column_list)
# extracting columns for numerical columns
valid_cols = [column for column in null_column_list if df[column].dtype != 'object']
# extracting columns for categorical columns
categorical_cols = [column for column in null_column_list if df[column].dtype == 'object']
numeric_cols = valid_cols
df_stats_mean = df.copy()
# Imputing mean for numeric values and then imputing mean for categorical values
print(f'Imputing following columns with mean : {numeric_cols}')
print(f'Imputing following columns with mode : {categorical_cols}')

if len(numeric_cols):
    for i in numeric_cols:
        df_stats_mean.fillna({i : df[i].mean()}, inplace=True)

if len(categorical_cols):
    for i in categorical_cols:
        df_stats_mean.fillna({i : df[i].mode()[0]}, inplace=True)

#return df_stats_mean, df_stats_median, df_stats_mode
```
