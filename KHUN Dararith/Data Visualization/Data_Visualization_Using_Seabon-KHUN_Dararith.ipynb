Seaborn harnesses the power of matplotlib to create beautiful charts in a few lines of code. ## <center>1. Quick Introduction 
Seaborn is an amazing visualization library for statistical graphics plotting in Python. It provides beautiful default styles and color palettes to make statistical plots more attractive.

Seaborn aims to make visualization the central part of exploring and understanding data.

### 1.1 Seabon Dist Plot
Seaborn dist plot  is used to plot a histogram, with some other variations like kdeplot and rugplot.

Here is an example:

```python
import numpy as np
import seaborn as sns

# Selecting style as white,
# dark, whitegrid, darkgrid 
# or ticks
sns.set(style="white")
  
# Generate a random univariate 
# dataset
rs = np.random.RandomState(10)
d = rs.normal(size=100)
  
# Plot a simple histogram and kde 
# with binsize determined automatically
sns.distplot(d, kde=True, color="m")
```

### 1.2 Seabon Line Plot
This plot is mainly used to visualize the data in form of some time series, i.e. in continuous manner.

Example: 

```python
import seaborn as sns
  
sns.set(style="dark")
fmri = sns.load_dataset("fmri")
  
# Plot the responses for different\
# events and regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
```

### 1.3 Seabon Lmplot
This plot shows a line representing a linear regression model along
with data points on the 2D-space and x and y can be set as the
horizontal and vertical labels respectively.

Example: 

```python
import seaborn as sns
  
sns.set(style="ticks")
  
# Loading the dataset
df = sns.load_dataset("anscombe")
  
# Show the results of a linear regression
sns.lmplot(x="x", y="y", data=df)
```

## <center>3. Distribution Plot
Distribution plots in seaborn which is used for examining 
univariate and bivariate distributions. 
In this article we will be discussing 4 types of
distribution plots namely:

1. Join plot
2. Dist plot
3. Pair plot
4. Rug plot

### 3.1 Join Plot
It is used to draw a plot of two variables with bivariate
and univariate graphs. It basically combines two different
plots.

Examples of Join Plot:

```python
sns.jointplot(x ='total_bill', y ='tip', data = df)
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = df, kind ='kde')
# KDE shows the density where the points match up the most
```

### 3.2 Dist Plot
It is used basically for univariant set of observations and 
visualizes.

Example of Dist Plot:

```python
# set the background style of the plot
sns.set_style('whitegrid')
sns.distplot(df['total_bill'], kde = False, color ='red', bins = 30)
```

### 3.3 Pair Plot
It represents pairwise relation across the entire 
dataframe and supports an additional argument called
hue for categorical separation. What it does basically
is create a jointplot between every possible numerical
column and takes a while if the dataframe is really huge.

Example of Pair Plot:

```python
sns.pairplot(df, hue ="sex", palette ='coolwarm')
```

### 3.4 Rug Plot
It plots datapoints in an array as sticks on an axis.
Just like a distplot it takes a single column. 
Instead of drawing a histogram it creates dashes 
all across the plot.

Example of Rug Plot: 

```python
sns.rugplot(df['total_bill'])
```

## <center>4. Regression Plot
The regression plots in seaborn are primarily intended to add a visual guide 
that helps to emphasize patterns in a dataset during exploratory data analyses.
Regression plots as the name suggests creates a regression line between 2 parameters
and helps to visualize their linear relationships.

### - Simple Linear Plot:

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = df)
```

### - Linear Plot with additional parameters:

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = df, 
           hue ='sex', markers =['o', 'v'])
```

### - Setting the size and color of the plot:

```python
sns.set_style('whitegrid')
sns.lmplot(x ='total_bill', y ='tip', data = df, hue ='sex', 
           markers =['o', 'v'], scatter_kws ={'s':100}, 
           palette ='plasma')
```

### - Displaying multiple plots:

```python
sns.lmplot(x ='total_bill', y ='tip', data = df, col ='sex', row ='time', hue ='smoker')
```

### - Size and aspect ratio of plots:

```python
sns.lmplot(x ='total_bill', y ='tip', data = df, col ='sex', 
           row ='time', hue ='smoker', aspect = 0.6, 
           size = 4, palette ='coolwarm')
```

## <center>6. Mutiple Plots
It is a useful approach to draw multiple instances of the same plot on different subsets of your dataset.
It allows a viewer to quickly extract a large amount of information about a complex dataset. 
In Seaborn, we will plot multiple graphs in a single window in two ways.
First with the help of Facetgrid() function and other by implicit with the help of matplotlib.


### - Faceet Grid: is a general way of ploting grids based on a function. 
It helps in visualizing distribution of one variable as well as the relationship 
between multiple variables. Its object uses the dataframe as Input and the names
of the variables that shape the column, row, or color dimensions of the grid

```python
# Form a facetgrid using columns
sea = sns.FacetGrid(exercise, col = "time")
```

```python
# Form a facetgrid using columns with a hue 
sea = sns.FacetGrid(exercise, col = "time", hue = "kind")
    
# map the above form facetgrid with some attributes   
sea.map(sns.scatterplot, "pulse", "time", alpha = .8)
  
# adding legend
sea.add_legend()
```

#### + Several options for controlling the look of the grid that can be passed to the class constructor.

```python
sea = sns.FacetGrid(exercise, row = "diet",
                    col = "time", margin_titles = True)
  
sea.map(sns.regplot, "id", "pulse", color = ".3",
        fit_reg = False, x_jitter = .1)
```

#### + Size of the figure is set by providing the height of each facet, along with the aspect ratio:


```python
sea = sns.FacetGrid(exercise, col = "time",
                    height = 4, aspect  =.5)
  
sea.map(sns.barplot, "diet", "pulse",
        order = ["no fat", "low fat"])
```

#### + The facets will be in the order of appearance of the category levels


```python
exercise_kind = exercise.kind.value_counts().index
sea = sns.FacetGrid(exercise, row = "kind", 
                    row_order = exercise_kind,
                    height = 1.7, aspect = 4)
  
sea.map(sns.kdeplot, "id")
```

#### + We can plot it along the columns but wrap them so that they span multiple rows

```python
g = sns.PairGrid(exercise)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```
